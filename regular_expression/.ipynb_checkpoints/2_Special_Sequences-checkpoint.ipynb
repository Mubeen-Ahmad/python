```python
import re
```

# Metacharacters
## Metacharacters are characters with a special meaning

# [ ]
* # A set of characters

```python
txt = "a0b1c2d3e4f5g6h7i8j9k10l"

# extract only d to f
re.findall("[d-f]",txt)
```

#  \ 	
* ## Signals a special sequence (can also be used to escape special characters)

### \d Digits
### \w symboles
### \W Letters and Numbers

```python
txt = "Hell0 World"

# FIND ALL DIGIT 
print(re.findall("\d",txt))

# Find symbole
print(re.findall("\W","Sweeterror404@gmai l.com"))

# FIND only letters and numbers
print(re.findall("\w","Sweeterror404@gmai l.com"))
```

# .
## Any character (except newline character)

```python
# find specfic character with after any 1 character
print(re.findall("e.","my name is Mubeen"))

# find specfic character with after any 2 character
print(re.findall("e..","my name is Mubeen"))

# find specfic character with before any 2 character
print(re.findall(".e","my name is Mubeen"))
```

#  ^ 	
# Starts with

```python
# Start with Hello
print(re.findall("^Hello","Hello World"))
```
