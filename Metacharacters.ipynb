# Metacharacters
## Metacharacters are characters with a special meaning

# [ ]
* # A set of characters

* ## Extract text from (d to f)

```python
import re
txt = "a0b1c2d3e4f5g6h7i8j9k10l"
re.findall("[d-f]",txt)
```

#  \ 	
* ## Signals a special sequence (can also be used to escape special characters)

    * ### \d Digits
    * ### \w symboles
    * ### \W Letters and Numbers
    * ### etc

```python
txt = "Hell0 World"

# FIND ALL DIGIT 
print(re.findall("\d",txt))

# Find symbole
print(re.findall("\W","Sweeterror404@gmai l.com"))

# FIND only letters and numbers
print(re.findall("\w","Sweeterror404@gmai l.com"))
```

# .
## Any character (except newline character)

```python
# find specfic character with any other 1 character
print(re.findall("e.","my name is Mubeen"))

# find specfic character with any other 2 characters
print(re.findall("e..","my name is Mubeen"))

# find specfic character with any other 2 characters
print(re.findall("..e","my name is Mubeen"))
```

#  ^ 	
# Starts with

```python
# Start with Hello
print(re.findall("^Hello","Hello World"))

# Start with We
print(re.findall("^We","We are anonmyous"))
```

# $
# Ends with

```python
# Ends with d
print(re.findall("d$","Hello World"))

# Ends with World
print(re.findall("World$","Hello World"))
```

# *
* # Zero or more occurrences

```python
txt = "ai aii aiii ia" 

# find all a and with all i
x = re.findall("ai*", txt)
print(x)
```

# + 
* ## One or more occurrences

```python
txt = "ai aii aiii ia" 

# find only 1 a and with all i
x = re.findall("ai+", txt)
print(x)
```

# ?
* # Zero or one occurrences

```python
txt = "ai aii aiii iaaoi" 

# find only 1 a and with all 1 i
x = re.findall("a.i?", txt)
print(x)
```
